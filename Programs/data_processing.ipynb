## Mapping Step + Concatenate Files:
- Define custom rules based on how we want to manage the updates - column mapping
- Define selected columns
- Create a function to apply these rules to the selected colums
- Concatenate prepared files

We have created a "test" folder for using some files for testing

```python
# 1. Define column mapping
# -----------------------------
COLUMN_MAP = {
    'USMCA_Eligible_Flag': 'USMCA Eligible Flag',
    'Net Available Inventory': 'Net Available Inventory Total',
    'Open Orders GDRC D': 'Open Orders GDRC-D',
    'ADD GDRC D Backorder': 'ADD GDRC-D Backorder',
    

}

# -----------------------------
# 2. Column selection for modeling
# -----------------------------
SELECTED_COLUMNS = [
    'Product ID', 'Brand', 'MFG Class', 'Resource ID', 'Franchise View', 'USMCA Eligible Flag', 'FDA FTZ Flag',
    'ABC Classification', 'Backorder/Backlog Total', 'Open Orders Total',
    'Net Available Inventory Total', 'LSP DOH Target', 'US DOH Target', 'Std Cost', 'ADD Total Incl. backorder'
]

# -----------------------------
# 3. Map function to unify column names
# -----------------------------
def unify_column_name(col):
    return COLUMN_MAP.get(col.strip(), col.strip())

# -----------------------------
# 4. Function to load and unify one file
# -----------------------------
def load_and_unify_excel(path):
    df = pd.read_excel(path, header=4)  # Asumes encabezado en fila 5
    df.rename(columns=lambda col: unify_column_name(col), inplace=True)
    df = df.dropna(how='all')  # Eliminar filas completamente vacías

    # Eliminar filas donde no hay Product ID y todas las columnas seleccionadas están vacías o nulas
    condition = df[SELECTED_COLUMNS].isna().all(axis=1)
    if 'Product ID' in df.columns:
        condition &= df['Product ID'].isna()
    df = df[~condition]

    for col in SELECTED_COLUMNS:
        if col not in df.columns:
            df[col] = pd.NA  # Rellenar con NA si falta
    return df[SELECTED_COLUMNS]

# -----------------------------
# 5. Main script to concatenate
# -----------------------------
files_path = "./test"

def concatenate_psr_files(files_path):
    dfs = []
    for file in sorted(os.listdir(files_path)):
        if file.endswith('.xlsx'):
            full_path = os.path.join(files_path, file)
            df = load_and_unify_excel(full_path)
            df['week'] = os.path.splitext(file)[0]
            dfs.append(df)
    final_df = pd.concat(dfs, ignore_index=True)
    final_df['week'] = pd.to_datetime(final_df['week'], errors='coerce')
    return final_df
```

```python
concatenate_psr_files(files_path)
```

Files are saved on "results" folder

```python
output_path = os.path.join("./results", 'psr_full_dataset.xlsx')
concatenate_psr_files(files_path).to_excel(output_path, index=False)
```
